```python
datasets = pd.read_csv('./breast_2_train.csv')
X_train = datasets.loc[:, datasets.columns != "y"]
X_train=X_train.loc[:,X_train.columns!="idx"]
y_train = datasets["y"]

X=th.tensor(np.array(X_train),dtype=th.float32)
y=th.tensor(np.array(y_train),dtype=th.float32)

# ÊääÊ†áÁ≠æËΩ¨‰∏∫2Áª¥
y = y.view(y.shape[0],1)
```
