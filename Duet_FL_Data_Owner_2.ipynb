# Part 1: Launch a Duet Server and upload data

```python
#引入相关依赖
from sklearn import datasets
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import torch as th
import torch.nn as nn
import pandas as pd
import numpy as np
import warnings
#忽略警告
warnings.filterwarnings('ignore')
import syft as sy
duet = sy.launch_duet(loopback=True,network_url="http://localhost:5000/")
```

```python
datasets = pd.read_csv('./breast_2_train.csv')
X_train = datasets.loc[:, datasets.columns != "y"]
X_train=X_train.loc[:,X_train.columns!="idx"]
y_train = datasets["y"]

X=th.tensor(np.array(X_train),dtype=th.float32)
y=th.tensor(np.array(y_train),dtype=th.float32)

# 把标签转为2维
y = y.view(y.shape[0],1)
```

```python
X = X.tag("breast_cancer-data")
y = y.tag("breast_cancer-target")

X = X.describe(
    "这是威斯康星洲临床科学中心开元的乳腺癌肿瘤数据集总296条样本，共30维的特征数据"
)
y = y.describe("这是数据标签，1代表良性肿瘤，0代表恶性肿瘤")

data_pointer = X.send(duet, searchable=True)
target_pointer = y.send(duet, searchable=True)

```
